#### [2.1.4] Use OpenAI API to categorize these unique jobs into categories

```python
import os
import openai
from dotenv import load_dotenv
```

```python
# Load environment variables from .env files
load_dotenv()
```

```python
# Set OpenAI API key
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
def categorize_job(job_title):
    """
    Categorizes a job title into one of 10 predefined broad categories using OpenAI's GPT.

    Parameters:
        job_title (str): The job title to categorize

    Returns: 
        str: The category of the job title
    """
    # Define the 11 predefined categories
    categories = [
        "Technology",
        "Healthcare",
        "Education",
        "Finance",
        "Engineering",
        "Science and Research",
        "Sales and Marketing",
        "Arts and Design",
        "Construction",
        "Administrative Support",
        "Other"
    ]
    
    try:
        # Create a formatted prompt with categories
        prompt = (
            f"Categorize the following job title into one of the following broad categories:\n"
            f"{', '.join(categories)}.\n\n"
            f"Job Title: {job_title}\n\n"
            f"Broad Category:"
        )
        
        # Make API call to OpenAI
        response = openai.Completion.create(
            model="gpt-3.5-turbo-instruct",
            prompt=prompt,
            temperature=0.3,
            max_tokens=10  # Limit response to a short category name
        )
        
        # Extract and clean up response text
        category = response['choices'][0]['text'].strip()
        
        # Ensure the category is valid (fallback to "Other" if invalid)
        if category not in categories:
            category = "Other"
        
        return category
        
    except Exception as e:
        print(f"Error categorizing job: {e}")
        return "Other"
```

In the OpenAI API, the temperature parameter controls the randomness or creativity of the model's responses. It determines how "creative" or "focused" the output is by influencing the sampling process when generating text.

```python
df['job_category'] = df['job'].apply(categorize_job)
```

#### [2.1.5] Covert dob into Age

```python
from datetime import datetime
```

```python
def calculate_age(dob):
    """
    Calculate age from date of birth.

    Parameters:
        dob (str): Date of birth in the format 'YYYY-MM-DD'.

    Returns:
        int: Age in years.
    """
    try:
        # Parse the DOB string into a datetime object
        dob_date = datetime.strptime(dob, "%Y-%m-%d")
        
        # Get today's date
        today = datetime.today() # Can change based on year of dataset
        
        # Calculate age
        age = today.year - dob_date.year - ((today.month, today.day) < (dob_date.month, dob_date.day))
        
        return age
    except ValueError as e:
        print(f"Invalid date format: {e}")
        return None
```

```python
df_copy['age'] = df_copy['dob'].apply(calculate_age) 
```
