### 1. Import Packages

```python
import os
import pandas as pd
```

```python
# Specify the directory containing the CSV files
directory = "../../data/raw/final"
```

```python
# Initialize an empty list to store DataFrames
dataframes = []
```

```python
# Iterate through files
for filename in os.listdir(directory):
    # Check to see if files start with 'tr' and is 'csv' file
    if filename.startswith("tr") and filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        # Read the CSV file and append it to the list
        dataframes.append(pd.read_csv(filepath,  delimiter="|"))
# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)
```

```python
merged_df
```

```python
output_file = os.path.join("../../data/interim/", "merged_output.csv")
merged_df.to_csv(output_file, index=False)

print(f"All CSV files starting with 'tr' have been merged and saved to {output_file}.")
```
